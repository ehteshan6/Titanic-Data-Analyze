## Matplotlib Excercise

```python
sns.barplot(x='Survived',y='Deaths',data=df)
plt.show();
```

```python
df['Death'] = 1 - df['Survived']  # Deaths = 1 when Survived = 0, otherwise 0

# Aggregating Deaths and Survivals
survival_counts = df.groupby('Survived').size().reset_index(name='Count')

# Bar Plot
sns.barplot(x=survival_counts['Survived'], y=survival_counts['Count'],palette='Set1')
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Number of Survivors and Deaths")
plt.show()
```
